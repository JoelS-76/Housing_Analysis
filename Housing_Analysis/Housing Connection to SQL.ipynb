```python
# Import dependencies
import os
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
# importing housing data
house_df = pd.read_excel('Working Data/clean_House_prices.xlsx')
house_df.head()
```

```python
# importing income data
income_df = pd.read_excel('Working Data/CO_income_clean.xlsx')
income_df
```

```python
# Read in csv file
pricing_out_df = pd.read_csv('Working Data/pricing_out.csv')
pricing_out_df.head()
```

```python
# Connection Strong to local server and database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Housing-Database"

#postgresql//postgres:{db_password}@localhost:5432/Housing-Database
#"postgresql://[user]:[password]@[location]:[port]/[database]"
```

```python
# Create database engine
engine = create_engine(db_string)
```

```python
# Import housing data 
house_df.to_sql(name='Housing', con=engine)
```

```python
# Import income data 
income_df.to_sql(name='Income', con=engine)
```

```python
# Import pricing out data 
pricing_out_df.to_sql(name='Pricing Out', con=engine)
```
