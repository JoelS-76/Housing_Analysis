```python
import pandas as pd
import json
```

```python
counties=open("counties.json")
counties = json.load(counties)
counties
```

```python
prices=open("priced_out.json")
prices = json.load(prices)
len(prices)
```

```python
len(counties["features"])
```

```python
count = 0
new_features = []
for p in prices:
    county = p["county"]
    for f in counties["features"]:
        print(f["properties"])
        name = f["properties"]["name"]
        if county == name:
            new_dict = {**f}
            new_dict["properties"]=p
            new_features.append(new_dict)
            #print (f"{count} county:{county}, name:{name}")
    count += 1
```

```python
new_features
```

```python
counties["features"]=new_features
counties
```

```python
with open ("clean_counties.json","w")as output:
    json.dump(counties, output)
```
